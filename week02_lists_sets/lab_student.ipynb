## Task 5 – Introduction to Sets

Create a list containing the following IP addresses:

192.168.1.1  
192.168.1.2  
192.168.1.1  
10.0.0.5  
10.0.0.5  

Convert this list into a **set**.

Print the set.

```python
ip_list = ["192.168.1.1","192.168.1.2","10.0.0.5","192.168.1.1","10.0.0.5"]
ip_list = set(ip_list)
print(ip_list)
```

## Task 6 – Unique Values

Using the set created in the previous task:

- Print each IP address on a separate line
- Print how many unique IP addresses exist

```python
ip_list = ["192.168.1.1","192.168.1.2","10.0.0.5","192.168.1.1","10.0.0.5"]
ip_list = set(ip_list)
counter = 0
for item in ip_list:
  counter += 1
  print(item)
u_ip_number = f"The number of unique IP addresses is {counter}"
print(u_ip_number)
```

## Task 7 – User Input with Lists

Ask the user to enter **five IP addresses**.

Store them in a list.

Then:

- Convert the list into a set
- Print the unique IP addresses
- Print the total number of unique IP addresses

```python
counter = 0
ip_list = []
while counter < 5:
  ask_ip = str(input("Enter an IP address : "))
  ip_list.append(ask_ip)
  counter += 1
make_unique = set(ip_list)
text = f"The total number of the unique IP addresses is {len(make_unique)}"
print(text , f"-> {make_unique}")
```
